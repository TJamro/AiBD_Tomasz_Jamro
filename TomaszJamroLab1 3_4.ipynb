```python
import matplotlib.pyplot as plt
import numpy as np

def fun(x):
    return (x**2)+5
plt.figure(1)
x1 = np.linspace(-1, 1, 100)
plt.plot(x1, fun(x))
plt.figure(2)
x2 = np.linspace(-6, 6, 100)
plt.plot(x2, fun(x))
plt.figure(3)
x3 = np.linspace(0, 5, 100)
plt.plot(x3, fun(x))
```
